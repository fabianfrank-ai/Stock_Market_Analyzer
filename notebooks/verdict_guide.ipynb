# Verdict system : categorical scoring
For this example we will take random indicator-results, to make you familiar with the verdict system. Even though I explained it thorougly in my code, I hope to make it easier to understand here. It uses the indicators and evaluates how good/bad they are. Depending on how good/bad they are, they will get a buy or sell signal, sometimes stronger with 3, sometimes weaker with 1.

For this example we will use following data:
- sma_diff (difference between long and short sma): 6

```python
    if sma_diff.iloc[-1] > 2 and sma_diff.iloc[-1] < 8:
        buy_signals += 1
    elif sma_diff.iloc[-1] < -2:
        sell_signals += 1
    elif sma_diff.iloc[-1] >= 10:
        buy_signals += 3
    elif sma_diff.iloc[-1] <= -10:
        sell_signals += 3
```

as we see, an sma_diff of six would give us a buy_signal of 1: we note: +1

- bollinger_percentage(where is the price between upper and lower band?): 0.9

```python
    if bollinger_percentage < 0.2:
        sell_signals += 3
    elif bollinger_percentage >= 0.2 and bollinger_percentage <= 0.4:
        sell_signals += 1
    elif bollinger_percentage >= 0.4 and bollinger_percentage <= 0.6:
        buy_signals += 1
    elif bollinger_percentage > 0.8:
        buy_signals += 3
```

A bollinger percentage of 0.9 would give us 3 buy signals: we note: 1+3=4
- rsi(explained in the indicators): 32

```python
    if rsi.iloc[-1] > 80:
         sell_signals += 3
    elif rsi.iloc[-1] > 60 and rsi.iloc[-1] <= 80:
        sell_signals += 1
    elif rsi.iloc[-1] >= 20 and rsi.iloc[-1] <= 40:
        buy_signals += 1
    elif rsi.iloc[-1] < 20:
        buy_signals += 3
    else:
        pass
```

An rsi of 32 leaves us with 3 buy signals: we note: 4+3=7

- ema_percentage(difference between long and short ema): 4


```python
    if ema_percentage.iloc[-1] > 2 and ema_percentage.iloc[-1] < 8:
        buy_signals += 1
    elif ema_percentage.iloc[-1] < -2:
        sell_signals += 1
    elif ema_percentage.iloc[-1] >= 10:
        buy_signals += 3
    elif ema_percentage.iloc[-1] <= -10:
        sell_signals += 3
    else:
        pass
     # No signal from EMA
```

An ema percentage score of 4 would leave us with one signal: 7+1 = 8

- macd_scale(the scaled difference between signal line and macd line): -12

```python

    if macd_scale.iloc[-1] > 2 and macd_scale.iloc[-1] < 9:
        buy_signals += 1
    elif macd_scale.iloc[-1] < -2 and macd_scale.iloc[-1] > -9:
        sell_signals += 1
    elif macd_scale.iloc[-1] >= 10:
        buy_signals += 3
    elif macd_scale.iloc[-1] <= -10:
        sell_signals += 3
    else:
        pass
```

A macd_scaled of -12 would leave us with 3 sell signals 

So overall we would have about 8 buy signals and 3 sell signals. Looking at the verdict system this leaves us with: 

```python
    # return verdict
    if buy_signals >= 6 and buy_signals < 11:
        return "Buy"
    if buy_signals >= 12:
        return "Strong Buy"
    elif sell_signals >= 6 and sell_signals < 11:
        return "Sell"
    elif sell_signals >= 12:
        return "Strong Sell"
    else:
        return "Hold"
```

a signal of 'Buy'
